```python
eigen_value, eigen_vector = np.linalg.eig(X_train.cov())
eigen_value = np.real(eigen_value)
eigen_vector = np.real(eigen_vector)
```

## Ex 1.5

Following the observations of the question (Q4), apply the method "PCA before OLS" that consists in applying OLS with Y and XtrainVp1:60q, where Vp1:60q contains the eigenvectors (associated with the 60 largest eigenvalues) of the covariance matrix. Run linear regression (with intercept), then plot the values of the coefficients (but not for the intercept). On another graph, do the same using the classical OLS.

___

```python
eigen_vector
```

```python
# the indice of the 60 largest eigenvalues
ind = np.argsort((eigen_value))[-60:]

# choose the associed eigenvectors
eigen_vector[ind].shape
```

```python
X_train_PCA = np.dot(X_train, (eigen_vector[ind]).transpose())
```

```python
X_train_PCA
```

```python
X_train_PCA.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X_train_PCA, y_train)
```

```python
print(reg.coef_)
```

```python
plt.scatter(range(60), reg.coef_)
```

```python
reg_without_PCA = LinearRegression().fit(X_train, y_train)
# print(reg_without_PCA.coef_)
plt.scatter(range(len(reg_without_PCA.coef_)), reg_without_PCA.coef_)
```

## Ex 1.6

Provide the intercept values for the 2 regressions from the previous question. Also, provide the mean value of the output variable Y (for the train set). Are the two intercepts equal? Comment. Exceptionally for this question, center and normalize the variables after PCA (the low dimensional ones). Run the regression and verify that the intercept is equal to the average of Y on the train set.

___

```python
print('intercept of regrssion after PCA: ', reg.intercept_)
print('intercept of regrssion without PCA: ', reg_without_PCA.intercept_)
```

```python
y_train.mean()
```

No, they are not equal. Because the trainning set is not centered.

```python
X_train_PCA_centered_norm = X_train_PCA.copy()
for i in range(X_train_PCA.shape[1]):
    X_train_PCA_centered_norm[:, i] = (
        X_train_PCA[:, i]-X_train_PCA[:, i].mean())/np.sqrt(X_train_PCA[:, i].var())
```

```python
reg_PCA_centered_norm = LinearRegression().fit(
    X_train_PCA_centered_norm, y_train)
# print(reg_without_PCA.coef_)
plt.scatter(range(len(reg_PCA_centered_norm.coef_)),
            reg_PCA_centered_norm.coef_)
```

```python
print('intercept of regrssion after PCA and standardization: ',
      reg_PCA_centered_norm.intercept_)
print(y_train.mean())
```

We can see that the intercept is equal to the average of Y on the train set.
