## Ex 1.8

Program the method of the forward variable selection. You can use the test statistics of the test for nullity (as seen during the course). For the moment, do not define the stop criterion for the method, i.e. add a variable at each time until all the variables are selected. Provide the order of the variable selection.

___

```python
import matplotlib.pyplot as plt
import matplotlib
from sklearn import linear_model
from scipy.stats import norm
from sklearn import datasets
%matplotlib inline
```

```python
# def forward_variable_selection(diabetes_X_learn, diabetes_y_learn):

diabetes_X_learn = X_train
diabetes_y_learn = y_train

diabetes_X_aug = np.column_stack(
    (np.ones((diabetes_X_learn.shape[0], 1)), diabetes_X_learn))

p = diabetes_X_aug.shape[1]
n = diabetes_X_aug.shape[0]

test = np.zeros((p, p))
pval_mem = np.zeros(p)
pval = np.zeros((p, p))
resids = diabetes_y_learn

var_sel = []
var_remain = list(range(p))
in_test = []


regr = linear_model.LinearRegression(fit_intercept=False)


for k in range(p):
    resids_mem = np.zeros((p, n))

    for i in var_remain:
        xtmp = diabetes_X_aug[:, [i]]
        regr.fit(xtmp, resids)

        # calcul de (x'x)
        xx = np.sum(diabetes_X_aug[:, [i]] ** 2)
        resids_mem[i, :] = regr.predict(xtmp) - resids
        sigma2_tmp = np.sum(resids_mem[i, :] ** 2) / xx
        test[k, i] = np.sqrt(n) * np.abs(regr.coef_) / (np.sqrt(sigma2_tmp))
        pval[k, i] = 2 * (1 - norm.cdf(test[k, i]))

    # separe en deux vecteurs la listes des variables séléctionnées et les autres
    best_var = np.argmax(test[k, :])
    var_sel.append(best_var)
    resids = resids_mem[best_var, :]
    pval_mem[k] = pval[k, best_var]
    var_remain = np.setdiff1d(var_remain, var_sel)

# return var_sel, pval_mem
```

```python
print("The order of the variable selection: ", var_sel)
```

## Ex 1.9 

Stop criterion: We choose to stop if the p-value is larger than 0.1. Illustrate the method providing (i) the 3 graphs of the test statistics obtained when selecting the 1st, 2nd and 3rd variables (in abscissa: the indices of the variables; in the ordinate: the value of the test statics), (ii) the graphs of the first 50 p-values (each associated to a selected variable). On the same plot, trace the horizontal line with the ordinate 0.1. Finally, provide the list of the selected variables.

___

```python
from scipy import stats
```

```python
def forward_variable_selection_cri(diabetes_X_learn, diabetes_y_learn):
    diabetes_X_aug = np.column_stack(
        (np.ones((diabetes_X_learn.shape[0], 1)), diabetes_X_learn))

    p = diabetes_X_aug.shape[1]
    n = diabetes_X_aug.shape[0]

    test = np.zeros((p, p))
    pval_mem = np.zeros(p)
    pval = np.zeros((p, p))
    resids = diabetes_y_learn

    var_sel = []
    var_remain = list(range(p))
    in_test = []

    regr = linear_model.LinearRegression(fit_intercept=False)

    for k in range(p):
        resids_mem = np.zeros((p, n))

        for i in var_remain:
            xtmp = diabetes_X_aug[:, [i]]
            regr.fit(xtmp, resids)

            # calcul de (x'x)
            xx = np.sum(diabetes_X_aug[:, [i]] ** 2)
            resids_mem[i, :] = regr.predict(xtmp) - resids
            sigma2_tmp = np.sum(resids_mem[i, :] ** 2) / xx
            test[k, i] = np.sqrt(n) * np.abs(regr.coef_) / \
                (np.sqrt(sigma2_tmp))
            pval[k, i] = 2 * (1 - norm.cdf(test[k, i]))

        # separe en deux vecteurs la listes des variables séléctionnées et les autres

        best_var = np.argmax(test[k, :])
        if (pval[k, best_var] > 0.1):
            return var_sel

        var_sel.append(best_var)
        resids = resids_mem[best_var, :]
        pval_mem[k] = pval[k, best_var]
        var_remain = np.setdiff1d(var_remain, var_sel)

    return var_sel
```

```python
var_sel_ = forward_variable_selection_cri(X_train, y_train)
```

## Ex 1.12

Apply the ridge regression method. For the choice of the regularization parameter, run the cross- validation on the “folds” defined in the previous question. Each “fold” is used to calculate the prediction risk while the resting ones are used for estimating the model. Then the 4 risks are averaged. Plot the estimated risk curve as a function of the regularization parameter (pay attention when choosing the range for the values of the regularization parameter). Provide the optimal regularization parameter and the corresponding risk for the test sample.
___

```python
from sklearn.linear_model import Ridge
```

```python
pred_errs = []
alpha_lg = np.array(np.linspace(-2, 4, 20))
alphas = 10**alpha_lg


# step 1
X_train_cv = np.concatenate((X_train_cv2, X_train_cv3, X_train_cv4), axis=0)
y_train_cv = np.concatenate((y_train_cv2, y_train_cv3, y_train_cv4), axis=0)
X_test_cv = X_train_cv1
y_test_cv = y_train_cv1

err = []
for alpha in alphas:
    regressor = Ridge(alpha)
    regressor.fit(X_train_cv, y_train_cv)
    prediction = regressor.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)

# step 1
X_train_cv = np.concatenate((X_train_cv1, X_train_cv3, X_train_cv4), axis=0)
y_train_cv = np.concatenate((y_train_cv1, y_train_cv3, y_train_cv4), axis=0)
X_test_cv = X_train_cv2
y_test_cv = y_train_cv2

err = []
for alpha in alphas:
    regressor = Ridge(alpha)
    regressor.fit(X_train_cv, y_train_cv)
    prediction = regressor.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)

# step 1
X_train_cv = np.concatenate((X_train_cv1, X_train_cv2, X_train_cv4), axis=0)
y_train_cv = np.concatenate((y_train_cv1, y_train_cv2, y_train_cv4), axis=0)
X_test_cv = X_train_cv3
y_test_cv = y_train_cv3

err = []
for alpha in alphas:
    regressor = Ridge(alpha)
    regressor.fit(X_train_cv, y_train_cv)
    prediction = regressor.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)

# step 1
X_train_cv = np.concatenate((X_train_cv1, X_train_cv2, X_train_cv3), axis=0)
y_train_cv = np.concatenate((y_train_cv1, y_train_cv2, y_train_cv3), axis=0)
X_test_cv = X_train_cv4
y_test_cv = y_train_cv4

err = []
for alpha in alphas:
    regressor = Ridge(alpha)
    regressor.fit(X_train_cv, y_train_cv)
    prediction = regressor.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)
```

```python
plt.semilogx(alphas, np.mean(pred_errs, axis=0), label="risk")
plt.xlabel("lambda")
plt.ylabel("prediction risk")
plt.title("Cross Validation")
plt.legend()
plt.show()
```

```python
index_min = (np.where(np.mean(pred_errs, axis=0) ==
                      min(np.mean(pred_errs, axis=0)))[0][0])
print("The optimal regularization parameter is ",
      alphas[index_min], "\nand the corresponding risk for the test sample is: ",
      np.mean(pred_errs, axis=0)[index_min])
```

```python
pred_errs = []
alpha_lg = np.array(np.linspace(-2, 8, 20))
alphas = 10**alpha_lg


# step 1
X_train_cv = np.concatenate((X_train_cv2, X_train_cv3, X_train_cv4), axis=0)
y_train_cv = np.concatenate((y_train_cv2, y_train_cv3, y_train_cv4), axis=0)
X_test_cv = X_train_cv1
y_test_cv = y_train_cv1

err = []
for alpha in alphas:
    clf = SVR(C=alpha)
    clf.fit(X_train_cv, y_train_cv)
    prediction = clf.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)

# step 2
X_train_cv = np.concatenate((X_train_cv1, X_train_cv3, X_train_cv4), axis=0)
y_train_cv = np.concatenate((y_train_cv1, y_train_cv3, y_train_cv4), axis=0)
X_test_cv = X_train_cv2
y_test_cv = y_train_cv2

err = []
for alpha in alphas:
    clf = SVR(C=alpha)
    clf.fit(X_train_cv, y_train_cv)
    prediction = clf.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)

# step 3
X_train_cv = np.concatenate((X_train_cv1, X_train_cv2, X_train_cv4), axis=0)
y_train_cv = np.concatenate((y_train_cv1, y_train_cv2, y_train_cv4), axis=0)
X_test_cv = X_train_cv3
y_test_cv = y_train_cv3

err = []
for alpha in alphas:
    clf = SVR(C=alpha)
    clf.fit(X_train_cv, y_train_cv)
    prediction = clf.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)

# step 4
X_train_cv = np.concatenate((X_train_cv1, X_train_cv2, X_train_cv3), axis=0)
y_train_cv = np.concatenate((y_train_cv1, y_train_cv2, y_train_cv3), axis=0)
X_test_cv = X_train_cv4
y_test_cv = y_train_cv4

err = []
for alpha in alphas:
    clf = SVR(C=alpha)
    clf.fit(X_train_cv, y_train_cv)
    prediction = clf.predict(X_test_cv)

    # prediction error
    err.append(sum((y_test_cv - prediction)**2) / y_test_cv.size)
pred_errs.append(err)
```

```python
plt.semilogx(alphas, np.mean(pred_errs, axis=0), label="risk")
plt.xlabel("C")
plt.ylabel("prediction risk")
plt.title("Cross Validation")
plt.legend()
plt.show()
```

```python
index_min = (np.where(np.mean(pred_errs, axis=0) ==
                      min(np.mean(pred_errs, axis=0)))[0][0])
print("The optimal C parameter is ",
      alphas[index_min], "\nand the corresponding risk for the test sample is: ",
      np.mean(pred_errs, axis=0)[index_min])
```
