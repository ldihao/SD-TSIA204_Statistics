No, they are not equal. Because the trainning set is not centered.

```python
# Centralize and normalize the variables after PCA
X_train_PCA_centered_norm = X_train_PCA.copy()
for i in range(X_train_PCA.shape[1]):
    X_train_PCA_centered_norm[:, i] = (
        X_train_PCA[:, i]-X_train_PCA[:, i].mean())/np.sqrt(X_train_PCA[:, i].var())
```

```python
# run linear regression for centralized and normalized variables after PCA
reg_PCA_centered_norm = LinearRegression().fit(
    X_train_PCA_centered_norm, y_train)
# print(reg_without_PCA.coef_)
plt.scatter(range(len(reg_PCA_centered_norm.coef_)),
            reg_PCA_centered_norm.coef_)
plt.title("Parameters of OLS with PCA and centralization and normalization")
plt.xlabel("Parameters")
plt.ylabel("Value")
```

```python
print('Intercept of regrssion after PCA and standardization: ',
      reg_PCA_centered_norm.intercept_)
print("Average of Y: ", y_train.mean())
```

We can see that the intercept is equal to the average of Y on the train set.

Program the method of the forward variable selection. You can use the test statistics of the test for nullity (as seen during the course).
```python
from sklearn import linear_model
from scipy.stats import norm
from sklearn import datasets
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
def forward_variable_selection(diabetes_X_learn, diabetes_y_learn):
    diabetes_X_aug = np.column_stack( (np.ones( (diabetes_X_learn.shape[0], 1 )), diabetes_X_learn )) 

    p = diabetes_X_aug.shape[1]
    n = diabetes_X_aug.shape[0]

    test = np.zeros((p,p))
    pval_mem = np.zeros(p)
    pval = np.zeros((p,p))
    resids = diabetes_y_learn

    var_sel = []
    var_remain = list(range(p))
    in_test = []


    regr = linear_model.LinearRegression(fit_intercept = False)


    for k in range(p): 
        resids_mem = np.zeros((p,n))

        for i in var_remain:
            xtmp = diabetes_X_aug [:,[i]]
            regr.fit(xtmp , resids)

            #calcul de (x'x)
            xx = np.sum( diabetes_X_aug [:,[i]] ** 2 )    
            resids_mem [i,:] = regr.predict(xtmp) - resids
            sigma2_tmp = np.sum (resids_mem [i,:] ** 2) / xx
            test[k,i] = np.sqrt(n) * np.abs(regr.coef_) / (np.sqrt( sigma2_tmp )) 
            pval[k,i] =   2 * (1 - norm.cdf(test[k,i]))

        ####separe en deux vecteurs la listes des variables séléctionnées et les autres
        best_var = np.argmax(test[k,:])
        var_sel.append(best_var)
        resids = resids_mem[best_var,:]
        pval_mem [k] = pval[k,best_var]
        var_remain = np.setdiff1d(var_remain,var_sel)

    return var_sel
```

```python
def forward_variable_selection(X, y):
    S = []
    n,p = X.shape
    A = range(p)
    r = y
    test_statistic = np.zeros((p,p))
    for i in range(p):
#         print("iteration: ", i)
        # choose the most relevant variable each time
        t_max = 0
        index_max = -1
        for j in range(p):
            # compare different variable's test statistic
            if j not in S:
                S.append(j)
                tmp_X = np.concatenate((np.ones((n, 1)), X[S]), axis=1)
                tmp_X_t_X = np.dot(tmp_X.T, tmp_X)
                
                try:
                    tmp_X_t_X_inverse = np.linalg.inv(tmp_X_t_X)
                except np.linalg.LinAlgError:
                    # Not invertible. Skip this one.
                    print("not invertible")
                    pass
                else:
                    tmp_theta = tmp_X_t_X_inverse.dot(tmp_X.T).dot(r.T)
                    variance = np.var(r - tmp_X.dot(tmp_theta).reshape(-1))
                    temp = tmp_theta[i + 1] / np.sqrt(abs(variance * np.var(tmp_X)))
                    test_statistic[i][j] = abs(temp)
                    if i == 176:
                        print(temp)
                    if temp > t_max:
                        t_max = temp
                        index_max = j
                S.remove(j)
        if(index_max != -1):
            S.append(index_max)
#             print(index_max)
            tmp_X = np.concatenate((np.ones((n, 1)), X[S]), axis=1)
            tmp_X_t_X = np.dot(tmp_X.T, tmp_X)
            tmp_X_t_X_inverse = np.linalg.inv(tmp_X_t_X)
            tmp_theta = tmp_X_t_X_inverse.dot(tmp_X.T).dot(r.T)
            r = r - tmp_X.dot(tmp_theta).reshape(-1)
        else: 
            break
    return S, test_statistic
```
